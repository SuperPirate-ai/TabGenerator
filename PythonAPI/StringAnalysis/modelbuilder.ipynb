```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import json
```

```python


# Define the strings and corresponding file names
strings = ["Fat_E_all_notes", "A_all_notes", "D_all_notes", "G_all_notes", "B_all_notes", "thin_E_all_notes"]
file_names = [f"{s}" for s in strings]

# Read the data from the files
data = []
labels = []

for i, file_name in enumerate(file_names):
    if os.path.exists(file_name):
        with open(file_name, 'r', encoding='utf-8') as file:
            json_data = json.load(file)
            values = json_data[file_name]
            data.add(values)
            labels.add(i)

# Convert to numpy arrays
data = np.array(data).reshape(-1, 1)
labels = np.array(labels)


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

```

```python
# Build the model
model = Sequential([
    Dense(64, input_shape=(1,), activation='relu'),
    Dense(64, activation='relu'),
    Dense(6, activation='softmax')
])

```

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=8, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Model accuracy: {accuracy * 100:.2f}%")
```

```python

# Save the model
model.save('string_classifier_model.h5')
```
